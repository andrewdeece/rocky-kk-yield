# Figures Reproduction & Appendix Derivations

This notebook regenerates all figures from the paper and reproduces the key SymPy derivations from **Appendix A** (non-linear KK reduction + β=2) and **Appendix B** (entropic-holographic coupling).

```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
import sympy as sp
from pathlib import Path
from IPython.display import Image, display

print("✅ Environment ready for full reproduction.")
```

## 1. Load & Display Main Figures

```python
fig_dir = Path("../figures")
for fig in ["fig1_kk_projection.png", "fig2_frequency.png", "fig3_multi_model.png"]:
    path = fig_dir / fig
    if path.exists():
        display(Image(filename=path))
    else:
        print(f"Missing figure: {fig}")
```

## 2. Appendix A: Non-Linear KK Reduction — β=2 Derivation

```python
# SymPy derivation of β=2 power-law suppression (Appendix A)
n, omega, R5, Lambda = sp.symbols('n omega R5 Lambda', positive=True)

# Integral for KK-mode contribution
integral_expr = sp.integrate(n**2 / (n**2 + (omega * R5)**2), (n, 0, Lambda))
high_freq_scaling = sp.simplify(integral_expr / (1 / omega**2))

display(high_freq_scaling)
print("High-frequency scaling ~ 1/ω² → β = 2 confirmed")
```

## 3. Appendix B: Entropic-Holographic Coupling & Debye Form

```python
# Debye-like response from entropic variation
omega = sp.symbols('omega', positive=True)
tau_ent, alpha0 = sp.symbols('tau_ent alpha0', positive=True)

alpha = alpha0 / (1 - sp.I * omega * tau_ent)
Re_alpha = sp.re(alpha)
Im_alpha = sp.im(alpha)

display(Re_alpha)
display(Im_alpha)
print("Debye form α(ω) reproduced from δS_ent/δφ")
```

## 4. Regenerate Figure 2 — Frequency Dependence

```python
# Reproducible Figure 2
omega_vals = np.logspace(-7, 2, 1000)
tau_ent_val = 0.01
alpha0_val = 1.0

real_part = alpha0_val / (1 + (omega_vals * tau_ent_val)**2)
imag_part = alpha0_val * omega_vals * tau_ent_val / (1 + (omega_vals * tau_ent_val)**2)

plt.figure(figsize=(10, 6))
plt.semilogx(omega_vals, real_part, 'b-', label='Re[α_eff]')
plt.semilogx(omega_vals, imag_part, 'r--', label='Im[α_eff]')
plt.axvline(1/tau_ent_val, color='green', linestyle='--', label='Artemis window crossover')
plt.xlabel('Frequency ω (Hz)')
plt.ylabel('Screening Function α_eff')
plt.title('Frequency Dependence of Screened Dilaton Coupling (Reproducible)')
plt.legend()
plt.grid(True)
plt.show()
```
